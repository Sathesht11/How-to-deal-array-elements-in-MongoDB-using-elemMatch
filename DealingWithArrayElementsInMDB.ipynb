2. Storelocation should not be in ‘Seattle’,and 'Newyork'
3. Customer gender male or female and include null values
4. Age limits between 20-30

```python
# We can use the $elemMatch in the query part of the find() function 
# to satisfy multiple conditions in the same array

sales_data=[]
for i in collection.find({'$and':[{'items':{'$elemMatch':{'name':'laptop','price':{'$lt':600}}}},
                        {'storeLocation':{'$nin':['Seattle','New York']}},
                        {'$or':[{'customer.gender':{'$in':['M','F']}},
                        {'customer.gender':{'$type':10}}]},{'customer.age':{'$gte':20,'$lte':30}},
                        {'purchaseMethod':'In store'}]},{'_id':0}):
  sales_data.append(i)
  
sales_data
```

```python
# Convert the list of docs into DataFrame using pandas package
import pandas as pd
df=pd.DataFrame(sales_data)
df
```
